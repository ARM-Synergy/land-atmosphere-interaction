```python
import act
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import pandas as pd
```

```python
#Read excel file
data = pd.read_csv('/data/home/jlau/land-atmosphere-interaction/notebooks/eddypro_43m_fluxnet_2025-05-14T083625_adv.csv', parse_dates=True, index_col='TIMESTAMP_START')
```

```python
filtered_data = data.replace(-9999, np.nan)
```

```python
#set plot size
plt.figure(figsize=(16,8))

#Plot the different plots
filtered_data['FC'].plot(color='blue')
filtered_data['FCH4'].plot(color='red')
filtered_data['LE'].plot(color='orange')
filtered_data['H'].plot(color='yellow')
```

```python
fig, axes = plt.subplots(4, 1, figsize=(16, 16))
filtered_data['FC'].plot(ax=axes[0], color='blue')
axes[0].set_ylabel('FC')
filtered_data['FCH4'].plot(ax=axes[1], color='purple')
axes[1].set_ylabel('FCH4')
filtered_data['LE'].plot(ax=axes[2], color='orange')
axes[2].set_ylabel('LE')
filtered_data['H'].plot(ax=axes[3], color='yellow')
axes[3].set_ylabel('H')
# Add vertical line at x=0 on all subplots
for ax in axes:
    ax.axvline(x=0, color='black', linestyle='--')
plt.tight_layout(rect=[0, 0, 1, 0.96])
fig.suptitle('Fluxes', size=16)
```
