```python
sensible_heat_flux = ds_ecor['sensible_heat_flux']
latent_flux = ds_ecor['latent_flux']
```

```python
LEH = sensible_heat_flux + latent_flux
RnG = NetR - G
```

```python
LEH_np = LEH.values
RnG_np = RnG.values
```

```python
# Flatten in case they are multidimensional
LEH_flat = LEH_np.flatten()
RnG_flat = RnG_np.flatten()

# Ensure both arrays have the same length
min_len = min(len(LEH_flat), len(RnG_flat))
LEH_clean = LEH_flat[:min_len]
RnG_clean = RnG_flat[:min_len]

# Remove NaNs and infinite values
mask = np.isfinite(LEH_clean) & np.isfinite(RnG_clean)
LEH_final = LEH_clean[mask]
RnG_final = RnG_clean[mask]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Linear regression
slope, intercept, r_value, p_value, std_err = linregress(RnG_final, LEH_final)
fit_line = slope * RnG_final + intercept

# 1:1 line range
min_val = min(RnG_final.min(), LEH_final.min())
max_val = max(RnG_final.max(), LEH_final.max())

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(RnG_final, LEH_final, alpha=0.6, label='Data')
plt.plot(RnG_final, fit_line, 'r', label=f'Regression: y = {slope:.2f}x + {intercept:.2f}\n$R^2$ = {r_value**2:.3f}')
plt.plot([min_val, max_val], [min_val, max_val], 'k--', label='1:1 Line')

# Labels and legend
plt.xlabel('Rn - G (W/m²)')
plt.ylabel('LE + H (W/m²)')
plt.title('LEH vs RnG with Regression and 1:1 Line')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
