```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Calculate variables
TE = ds_ecor['sensible_heat_flux'] + ds_ecor['latent_flux']
Rn = (ds_sirs['down_long_hemisp1'] + ds_sirs['down_short_hemisp']) - (ds_sirs['up_long_hemisp'] + ds_sirs['up_short_hemisp'])
G = ds_sebs['surface_soil_heat_flux_avg']
available_energy = Rn - G

# Align datasets on common coordinates (e.g., time)
TE, available_energy = xr.align(TE, available_energy, join='inner')

# Convert to NumPy arrays
x = available_energy.values
y = TE.values

# Remove NaNs
valid = ~np.isnan(x) & ~np.isnan(y)
x_clean = x[valid]
y_clean = y[valid]

# Linear regression
slope, intercept, r_value, p_value, std_err = linregress(x_clean, y_clean)

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(x_clean, y_clean, s=10, alpha=0.6, label='Data')
plt.plot(x_clean, intercept + slope * x_clean, color='red', linewidth=2, label=f'Fit: y = {slope:.2f}x + {intercept:.2f}')
plt.xlabel('Rn - G (W/m²)')
plt.ylabel('H + LE (W/m²)')
plt.title(f'Energy Balance Closure\nSlope = {slope:.2f}')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
from scipy.stats import linregress

# Compute energy terms
TE = ds_ecor['sensible_heat_flux'] + ds_ecor['latent_flux']
Rn = (ds_sirs['down_long_hemisp1'] + ds_sirs['down_short_hemisp']) - (ds_sirs['up_long_hemisp'] + ds_sirs['up_short_hemisp'])
G = ds_sebs['surface_soil_heat_flux_avg']
available_energy = Rn - G

# Align all arrays by time
TE, available_energy = xr.align(TE, available_energy, join='inner')

# Convert to DataFrame for easier time-based grouping
df = xr.Dataset({'TE': TE, 'AE': available_energy}).to_dataframe().dropna()
df = df.loc['2024-10-01':'2025-04-30']  # Restrict to the specified time range
df['month'] = df.index.to_period('M')

# Prepare containers for results
results = []

# Loop through each month
for month, group in df.groupby('month'):
    x = group['AE'].values
    y = group['TE'].values

    if len(x) > 2:  # Require at least 3 data points to fit
        slope, intercept, r, p, stderr = linregress(x, y)
        results.append({
            'month': month,
            'slope': slope,
            'r_squared': r**2
        })

# Convert results to DataFrame and sort chronologically
results_df = pd.DataFrame(results).sort_values('month')
month_labels = results_df['month'].dt.strftime('%b %Y')

# Plotting
fig, ax1 = plt.subplots(figsize=(10, 5))

# Slope line
ax1.plot(month_labels, results_df['slope'], marker='o', color='blue', label='Slope')
ax1.set_ylabel('Slope', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_xticklabels(month_labels, rotation=45)
ax1.set_title('Monthly Energy Balance Closure: Slope and R²')

# R² line on secondary y-axis
ax2 = ax1.twinx()
ax2.plot(month_labels, results_df['r_squared'], marker='s', color='green', label='R²')
ax2.set_ylabel('R²', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

plt.tight_layout()
plt.grid(True)
plt.show()
```
